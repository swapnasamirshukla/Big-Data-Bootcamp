### Q1. Why do we call Python as a general purpose and high-level programming language?
**Answer**: Python is called a general-purpose language because it's designed to be used in a wide variety of application domains, from web development to data science to robotics. It's high-level because it abstracts many of the details of computer operation, allowing programmers to focus on logic and functionality rather than low-level operations.

### Q2. Why is Python called a dynamically typed language?
**Answer**: Python is dynamically typed because variable types are determined at runtime, rather than being specified at compile-time. You don't have to explicitly state the type of a variable when you declare it; Python will figure it out for you.

### Q3. List some pros and cons of Python programming language?
**Pros**:
- Readable and maintainable code.
- Extensive standard libraries and community contributions.
- Supports multiple programming paradigms.
- Good for rapid development.

**Cons**:
- Slower execution speed compared to compiled languages like C++.
- Global interpreter lock (GIL) makes it less suitable for multi-threading.
- Higher memory consumption.

### Q8. What is type casting?
**Answer**: Type casting is the conversion of one data type to another, like converting an integer to a float, or a string to an integer. E.g., `int("10")` converts the string "10" to an integer.

### Q17. Write a code to take a number as an input from the user and check if the number is odd or even.
```python
number = int(input("Enter a number: "))
if number % 2 == 0:
    print("Even")
else:
    print("Odd")
```

### Q20. What are conditional statements in Python?
**Answer**: Conditional statements (`if`, `elif`, `else`) are used to perform different actions based on different conditions.

### Q21. What is the use of 'if', 'elif' and 'else' keywords?
**Answer**: `if` is used to test a condition; `elif` is used to test additional conditions if the initial `if` fails; `else` is used to specify what to do if all `if` and `elif` conditions fail.

### Q22. Write a code to take the age of a person as an input and if age >= 18 display "I can vote". If age is < 18 display "I can't vote".

### Q23. Write a code that displays the sum of all the even numbers from the given list.
```python
numbers = [12, 75, 150, 180, 145, 525, 50]
print(sum(x for x in numbers if x % 2 == 0))
```

### Q24. Write a code to take 3 numbers as an input from the user and display the greatest no as output.
```python
a, b, c = map(int, input("Enter three numbers separated by spaces: ").split())
print("Greatest:", max(a, b, c))
```

### Q25. Write a code to display only those numbers from a list that satisfy the specified conditions.

- The number must be divisible by five

- If the number is greater than 150, then skip it and move to the next number

- If the number is greater than 500, then stop the loop

```python
numbers = [12, 75, 150, 180, 145, 525, 50]
for num in numbers:
    if num > 500:
        break
    if num > 150:
        continue
    if num % 5 == 0:
        print(num)
```
